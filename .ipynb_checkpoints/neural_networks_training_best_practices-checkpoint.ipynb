### Training Neural Networks 

in this lesson we are about to learn the basic concepts to keep in mind when training ANNs

we will try to get an intuition as to the order of activities by which we are likely to get good results

and lay the foundations to train more complex and diverse networks in the comming practical lectures


we will use a very well known dataset called caltech-256 that can be downloaded [here](http://www.vision.caltech.edu/Image_Datasets/Caltech256/)

at first we will only use a subset of 20 categories out of (guess how many...) 257 categories :-)

this is ment to allow us iterating fastly through many concepts, in addition to loading all of our data to RAM

later we will learn about generators and use them to both train all 257 categories and enable easy augmentation of our training data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

```

```python
from keras.models import *
from keras.applications import VGG16
from keras.applications.resnet50 import ResNet50
from keras.layers import *
from keras.utils import to_categorical
from keras.callbacks import *
from keras.optimizers import *
from keras.models import *
from keras.layers import *
from keras.utils import to_categorical
from keras.callbacks import *

import os

import PIL

np.random.seed(2018)
checkpoint_string = 'bst_model_weights.h5'

n_folds = 5
```

```python
def read_and_resize_single_img(filename,nrow=224,ncol=224,channels=3):
    from scipy import misc
    arr = misc.imread(filename)
    arr = misc.imresize(arr,size=(nrow,ncol,channels))
    return arr

def read_train_from_dirs(num_imgs_per_class=5,skip=0,nrow=224,ncol=224,channels=3):
    import glob
    filelist = glob.glob('./256_ObjectCategories/*/*.jpg')
    categories = [l.split('/')[-2] for l in filelist]
    k2v = {i:l for i,l in enumerate(np.unique(categories))}
    v2k = {l:i for i,l in enumerate(np.unique(categories))}
    l = [[t,v2k[t]] for t in categories]
    labels_df = pd.DataFrame(l,columns=['category','cat_num'])
    labels_df['filenames'] = filelist
    labels_indexes = np.array([np.where(labels_df.cat_num==c)[0][skip:skip+num_imgs_per_class] 
                               for c in labels_df.cat_num.unique()[:20]])
    filtered = labels_df.iloc[labels_indexes.flatten(),:]
    num_imgs = filtered.shape[0]
    X = np.zeros((num_imgs,nrow,ncol,channels))
    for i in range(filtered.shape[0]):
        arr=read_and_resize_single_img(filtered.loc[:,'filenames'].iloc[i],nrow,ncol,channels)
        if len(arr.shape)<3: arr = np.stack([arr,arr,arr],axis=-1) #change to use np.repeat() for more generalizable
        X[i] = np.float16(arr)
    y = filtered.loc[:,'cat_num'].values
    return X,y,k2v,v2k
```

```python
def plot_multiple_imgs(data,nrows=4,ncols=4,figsize=(18,18)):
    fig,ax = plt.subplots(nrows=nrows,ncols=ncols,figsize=figsize)
    for i in range(nrows*ncols):
        ax[i%nrows,i//nrows].imshow(data[i])
```

```python
def save_model(model,filename):
    # this is a helper function used to save a keras NN model architecture and weights
    json_string = model.to_json()
    if not os.path.isdir('cache'):
        os.mkdir('cache')
    open(os.path.join('cache', filename+'_architecture.json'), 'w').write(json_string)
    model.save_weights(os.path.join('cache', filename+'_model_weights.h5'), overwrite=True)
    
def read_model(filename):
    # this is a helper function used to restore a keras NN model architecture and weights
    model = model_from_json(open(os.path.join('cache', filename+'_architecture.json')).read())
    model.load_weights(os.path.join('cache', filename+'_model_weights.h5'))
    return model
```

```python
def set_callbacks(description='run1',patience=3,tb_base_logdir='./logs/'):
    cp = ModelCheckpoint('best_model_weights_{}.h5'.format(description),save_best_only=True)
    es = EarlyStopping(patience=patience,monitor='val_acc')
    if not os.path.isdir('csv_logs'):
        os.mkdir('csv_logs')
    if not os.path.isdir('csv_logs/{}'.format(description)):
        os.mkdir('csv_logs/{}'.format(description))    
    log = CSVLogger('csv_logs/{}/train_log.csv'.format(description))
    tb = TensorBoard(log_dir='{}{}'.format(tb_base_logdir,description))
    cb = [cp,es,log,tb]
    return cb
```

```python
inp = Input(X_train.shape[1:])
x = Flatten()(inp)
x = Dense(20,activation='relu')(x)
x = Dense(to_categorical(y_train).shape[1],activation='softmax')(x)

model = Model(inputs=inp,outputs=x)
model.summary()
model.compile(loss='categorical_crossentropy',optimizer='adadelta',metrics=['accuracy'])
model.fit((X_train/255.0),to_categorical(y_train),
          validation_data=[(X_val/255.0),to_categorical(y_val)],epochs=30,
          callbacks=set_callbacks('fully_connected_1'))
```

### so we can easily notice that we don't get any improvement...
what seems to be the problem?

when we see that there is no change (at all) the first few epochs of a network it is usually due to one of the following:
1. data is not normlized so its hard for the network to converge
2. our optimizer is using too low learning rate (for too high we would notice diverging results)
3. we have somehow messed up eighter the training or validation data (or both)

lets try initially to use batch norm as a first layer to the model,and see if that helps...

```python
mean_0,mean_1,mean_2 = X_train[:,:,:,0].mean(),X_train[:,:,:,1].mean(),X_train[:,:,:,2].mean()
mean_0,mean_1,mean_2
```

```python
inp = Input(X_train.shape[1:])
x = BatchNormalization()(inp)
x = Flatten()(x)
x = Dense(20,activation='relu')(x)
x = Dense(to_categorical(y_train).shape[1],activation='softmax')(x)

model = Model(inputs=inp,outputs=x)
model.summary()
model.compile(loss='categorical_crossentropy',optimizer='adadelta',metrics=['accuracy'])
model.fit(X_train,to_categorical(y_train),validation_data=[X_val,to_categorical(y_val)],epochs=30,
          callbacks=set_callbacks('fully_connected_2_init_with_BN',patience=8))
```

### yes! our instincts were right and it was normalization that prevented us from fitting the data

lets try a bit wider net to see if we can fit a bit faster...

```python
inp = Input(X_train.shape[1:])
x = BatchNormalization()(inp)
x = Flatten()(x)
x = Dense(50,activation='relu')(x)
x = Dense(to_categorical(y_train).shape[1],activation='softmax')(x)

model = Model(inputs=inp,outputs=x)
model.summary()
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,to_categorical(y_train),validation_data=[X_val,to_categorical(y_val)],epochs=30,
          callbacks=set_callbacks('fully_connected_3_(wider)_init_with_BN',patience=8))
```

### so our wider net isn't overfitting as quickly as the more narrow one...
but as we can see its not fitting either.

lets check how will a deeper net do...

```python
inp = Input(X_train.shape[1:])
x = BatchNormalization()(inp)
x = Flatten()(x)
x = Dense(20,activation='relu')(x)
x = Dense(20,activation='relu')(x)
x = Dense(20,activation='relu')(x)
x = Dense(to_categorical(y_train).shape[1],activation='softmax')(x)

model = Model(inputs=inp,outputs=x)
model.summary()
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,to_categorical(y_train),validation_data=[X_val,to_categorical(y_val)],epochs=30,
          callbacks=set_callbacks('fully_connected_4_(deeper)_init_with_BN',patience=8))
```

### ok, so far we have tried fully connected networks and this has only got us to ~25% acc 
that is surely far from good result

and by now we know a thing or two about NN types 

so we should probably try CNN to use the connections between neering pixels in the images

```python
inp = Input((X_train.shape[1:]))
x = BatchNormalization()(inp)
x = Conv2D(64,(7),activation='relu')(x)
x = Conv2D(64,(3),activation='relu')(x)
x = Flatten()(x)
# x = Dense(20,activation='relu')(x)
x = Dense(to_categorical(y_train).shape[1],activation='softmax')(x)

model = Model(inputs=inp,outputs=x)
model.summary()
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,to_categorical(y_train),validation_data=[X_val,to_categorical(y_val)],epochs=3)
```

we can try to train such a net but we'll probably run out of gpu memory...

lets try a more simple one instead:

```python
inp = Input((X_train.shape[1:]))
x = BatchNormalization()(inp)
x = Conv2D(8,(7),activation='relu')(x)
x = MaxPooling2D(2)(x)
x = Flatten()(x)
# x = Dense(20,activation='relu')(x)
x = Dense(to_categorical(y_train).shape[1],activation='softmax')(x)

model = Model(inputs=inp,outputs=x)
model.summary()
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,to_categorical(y_train),validation_data=[X_val,to_categorical(y_val)],epochs=30,
         callbacks=set_callbacks('basic_cnn_init_with_BN',patience=8))
```

```python
inp = Input((X_train.shape[1:]))
x = BatchNormalization()(inp)
x = Conv2D(16,(7),activation='relu')(x)
x = Conv2D(8,(3),activation='relu')(x)
x = MaxPooling2D(2)(x)
x = Flatten()(x)
# x = Dense(20,activation='relu')(x)
x = Dense(to_categorical(y_train).shape[1],activation='softmax')(x)

model = Model(inputs=inp,outputs=x)
model.summary()
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,to_categorical(y_train),validation_data=[X_val,to_categorical(y_val)],epochs=30,
         callbacks=set_callbacks('2nd_cnn_init_with_BN',patience=8))
```

### great! we achived 99% acc on training set through 4 epochs only

```python
inp = Input((X_train.shape[1:]))
x = BatchNormalization()(inp)
x = Conv2D(16,(7),activation='relu')(x)
x = Conv2D(8,(3),activation='relu')(x)
x = Dropout(0.5)(x)
x = MaxPooling2D(2)(x)
x = Flatten()(x)
# x = Dense(20,activation='relu')(x)
x = Dense(to_categorical(y_train).shape[1],activation='softmax')(x)

model = Model(inputs=inp,outputs=x)
model.summary()
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,to_categorical(y_train),validation_data=[X_val,to_categorical(y_val)],epochs=30,
         callbacks=set_callbacks('3rd_cnn_init_with_BN_and_dropout',patience=8))
```

### by now we have only looked at exapmles of training our network from scratch
lets try and use some transfer learning from a well known architecture trained on imagenet called VGG16

we'll first just use fine tuning of the last model layer to see how that helps

notice that we set the 'trainable' attribute of all other layers to False

```python
X_train_norm = X_train/255.0 
X_val_norm = X_val/255.0 

from keras.applications.vgg16 import preprocess_input

vgg16 = VGG16(weights='imagenet', include_top=True)
vgg16.summary()
```

```python
for l in vgg16.layers:
    l.trainable = False

x = MaxPool2D(pool_size=(2),padding='valid')(vgg16.layers[-5].output)
x = Flatten()(x)
x = Dense(to_categorical(y_train).shape[1],activation='softmax',name = 'predictions')(x)
model = Model(inputs = vgg16.input,outputs = x)
model.summary()
model.compile(loss='categorical_crossentropy',optimizer='adadelta',metrics=['accuracy'])
model.fit(X_train_norm,to_categorical(y_train),validation_data=[X_val_norm,to_categorical(y_val)],epochs=150,
          callbacks=set_callbacks('224x224x3_vgg16_no_fc_layers',patience=30))
```

### data generators 
data generators are a powerful tool to add to our toolbox

they help us deal with several issues that we might face while tring to train ANNs:
1. data is too large to fit memory and we can not load it in advance
2. we would like to use data augmentation to reuse some of the data samples and help our models to generalize
3. we want to train our models using more than a single GPU


```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

def get_model_all_classes():
    vgg16 = VGG16(weights='imagenet', include_top=True)

    for l in vgg16.layers:
        l.trainable = False

    x = MaxPool2D(pool_size=(2),padding='valid')(vgg16.layers[-5].output)
    x = Dropout(0.3)(x)
    x = Flatten()(x)
    x = Dense(to_categorical(range(257)).shape[1],activation='softmax',name = 'predictions')(x)
    model = Model(inputs = vgg16.input,outputs = x)
    return model
```

```python
model = get_model_all_classes()

# its important to recreate the model so that we do not continue from where we stopped the last training 
model.compile(loss='categorical_crossentropy',optimizer='adadelta',metrics=['accuracy'])

batch_size = 32

datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

train_generator = datagen.flow_from_directory(
        directory = '256_ObjectCategories',  # this is the target directory
#         classes = [k2v[i] for i in np.unique(y_train)], #these are the valid classes
        target_size = (224, 224),  # all images will be resized to 224,224
        batch_size=batch_size,
        class_mode='categorical')

validation_generator = datagen.flow_from_directory(
        directory = '256_ObjectCategories',  # this is the target directory
#         classes = [k2v[i] for i in np.unique(y_train)], #these are the valid classes
        target_size = (224, 224),  # all images will be resized to 224,224
        batch_size=batch_size,
        class_mode='categorical')
```

```python
model.fit_generator(
        train_generator,
        steps_per_epoch=2000 // batch_size,
        epochs=50,
        validation_data=validation_generator,
        validation_steps=800 // batch_size)
```

```python
# its generaly a good practice to seperate train and validation into different folders so that they don't get mixed
num_train_samples, num_val_samples = 10,5


import glob
import shutil
import os

filelist = glob.glob('./256_ObjectCategories/*/*.jpg')
def create_train_val_sample_folder_structure(filelist,num_tr=10,num_val=5,verbose=False):
    if not os.path.isdir('./sample'): os.mkdir('./sample')
    if not os.path.isdir('./sample/train'): os.mkdir('./sample/train')
    if not os.path.isdir('./sample/validation'): os.mkdir('./sample/validation')
    file_df = pd.DataFrame({'category':[x.split('/')[-2] for x in filelist],'file_path':filelist})
    for c in file_df['category'].unique()[:]:
        if not os.path.isdir('./sample/validation/{}'.format(c)): os.mkdir('./sample/validation/{}'.format(c))
        if not os.path.isdir('./sample/train/{}'.format(c)): os.mkdir('./sample/train/{}'.format(c))
        for f in file_df['file_path'].loc[file_df['category']==c].iloc[:num_tr]:
            shutil.copy(f,'./sample/train/{}/{}'.format(f.split('/')[-2],f.split('/')[-1]))
            if verbose: print('copied file {}'.format(f))
        for f in file_df['file_path'].loc[file_df['category']==c].iloc[num_tr:num_tr+num_val]:
            shutil.copy(f,'./sample/validation/{}/{}'.format(f.split('/')[-2],f.split('/')[-1]))
            if verbose: print('copied file {}'.format(f))

create_train_val_sample_folder_structure(filelist,num_train_samples, num_val_samples)
```

```python
#this part is adjusted from françois chollet's tutorial

def save_bottlebeck_features(train_dir,val_dir,nrow=224,ncol=224,batch_size=5,save_to_disk=False):
    datagen = ImageDataGenerator(rescale=1. / 255)

    # build the VGG16 network
    model = VGG16(include_top=False, weights='imagenet')

    generator = datagen.flow_from_directory(
        train_dir,
        target_size=(ncol, nrow),
        batch_size=batch_size,
        class_mode=None,
        shuffle=False)
    bottleneck_features_train = model.predict_generator(
        generator, (num_train_samples // batch_size)*257)
    if save_to_disk: np.save(open('bottleneck_features_train.npy', 'w'),
            bottleneck_features_train)

    generator = datagen.flow_from_directory(
        val_dir,
        target_size=(ncol, nrow),
        batch_size=batch_size,
        class_mode=None,
        shuffle=False)
    bottleneck_features_validation = model.predict_generator(
        generator, (num_val_samples // batch_size)*257)
    if save_to_disk: np.save(open('bottleneck_features_validation.npy', 'w'),
            bottleneck_features_validation)
    return bottleneck_features_train,bottleneck_features_validation
```

```python
# this part is adjusted from a script by Yassine Alouini - Hyperopt the Xgboost model on kaggle.com

from hyperopt import STATUS_OK, Trials, fmin, hp, tpe
import xgboost as xgb

def score(params):
    print("Training with params: ")
    print(params)
    num_round = int(params['n_estimators'])
    del params['n_estimators']
    dtrain = xgb.DMatrix(X_train, label=y_train)
    dvalid = xgb.DMatrix(X_val, label=y_val)
    watchlist = [(dvalid, 'eval'), (dtrain, 'train')]
    gbm_model = xgb.train(params, dtrain, num_round,
                          evals=watchlist,
                          verbose_eval=True)
    predictions = gbm_model.predict(dvalid,
                                    ntree_limit=gbm_model.best_iteration + 1)
    score = roc_auc_score(y_valid, predictions)
    print("\tScore {0}\n\n".format(score))
    # The score function should return the loss (1-score)
    # since the optimize function looks for the minimum
    loss = 1 - score
    return {'loss': loss, 'status': STATUS_OK}


def optimize(
             #trials, 
             random_state=2018):
    """
    This is the optimization function that given a space (space here) of 
    hyperparameters and a scoring function (score here), finds the best hyperparameters.
    """
    # To learn more about XGBoost parameters, head to this page: 
    # https://github.com/dmlc/xgboost/blob/master/doc/parameter.md
    space = {
        'n_estimators': hp.quniform('n_estimators', 100, 300, 1),
        'eta': hp.quniform('eta', 0.025, 0.5, 0.025),
        # A problem with max_depth casted to float instead of int with
        # the hp.quniform method.
        'max_depth':  hp.choice('max_depth', np.arange(1, 9, dtype=int)),
        'min_child_weight': hp.quniform('min_child_weight', 1, 6, 1),
        'subsample': hp.quniform('subsample', 0.5, 1, 0.05),
        'gamma': hp.quniform('gamma', 0.5, 1, 0.05),
        'colsample_bytree': hp.quniform('colsample_bytree', 0.5, 1, 0.05),
        'eval_metric': 'mlogloss',
        'objective': 'multi:softprob',
        # Increase this number if you have more cores. Otherwise, remove it and it will default 
        # to the maxium number. 
        'nthread': 4,
        'booster': 'gbtree',
        'num_class':257,
        'tree_method': 'exact',
        'silent': 1,
        'seed': random_state
    }
    # Use the fmin function from Hyperopt to find the best hyperparameters
    best = fmin(score, space, algo=tpe.suggest, 
                # trials=trials, 
                max_evals=250)
    return best
```
